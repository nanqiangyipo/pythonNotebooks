```python
help(tf.reshape)
```

```python
x=tf.Variable(np.arange(9))
t=tf.reshape(x,(3,3))
init=tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(t))
    print(sess.run(tf.reshape(t,[-1,2,2])))
```
